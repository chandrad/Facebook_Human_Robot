```python
import pandas as pd
import pylab as pl
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import os 
```

```python
def scaledf(dfInp):
	#scale from 0 to 1
	return (dfInp - dfInp.min())/(dfInp.max() - dfInp.min())
```

```python
df = pd.read_csv("bidsReducedTrain.csv")
```

```python
features = [ "n_per_auc","n_devices", "n_countries", "n_ip", "n_url"]
```

```python
test_idx = np.random.uniform(0, 1, len(df)) <= 0.3
trainSet = df[test_idx==True] 
testSet = df[test_idx == False]

results = []
```

```python
for w in ['uniform', 'distance', lambda x: np.log(x)]:
	for n in range(1, 50, 1):
		clf = KNeighborsClassifier(n_neighbors=n,weights=w)
		clf.fit(trainSet[features], trainSet['outcome'])
		preds = clf.predict(testSet[features])
		accuracy = np.where(preds==testSet['outcome'], 1, 0).sum() / float(len(testSet))
		#accuracy = np.sqrt(np.sum((np.exp(preds) - np.exp(testSet['outcome']))**2)/float(len(testSet['outcome'])))
		print ("Neighbors: %d, Accuracy: %3f" % (n, accuracy))
		results.append([n, accuracy])
	results = pd.DataFrame(results, columns=["n", "accuracy"])
	pl.plot(results.n, results.accuracy)
	pl.title("Accuracy with Increasing K")
	pl.show()
```
